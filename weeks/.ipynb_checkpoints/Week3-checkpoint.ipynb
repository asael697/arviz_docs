## Report of the evaluated plots: week 2

The week 2 report contains the most relevant comments and issues found in the following plots:

 + Posterior
 
 + PPC
 
 + Rank
 
 + Separation
 
 + Trace
 
 + ts
 
 + Violin

For the evaluation I use `ArviZ: 0.13.0` version downloaded from Github.

## Relevant comments

This section provides the most relevant comments in every plot evaluation, followed by a tracking code using the defined issue when it is need.

### 1. Posterior plots

#### Analysis of current defaults

 + These plots should be the default option for the `plot_density` function as they have the same utility, but different aesthetics.
 
#### Analysis of usage advise

  + By defautl, set `skipna` argument to `False`.
  
#### Relevant Issues

 + Issue: [#2090](https://github.com/arviz-devs/arviz/issues/2080): Reorganize layout in Posterior Plot 

### 2. Posterior Predictive checks plots

#### Analysis of current defaults

 + The default `scatter` ppc plot provides a confusing visualization, both density and scatter points overlay without any distintion between one plot and the other.
 
 + This function has a significantly short functionality and diversity compared to the ppc methods provided by the bayesplot package.
 
#### Analysis of usage advise

 + By default, set the mean argument to `False`.
 
 + The label for the description for the `observed_rug` argument is missing. The `observed` argument is listed twice.
 
 + Update the **See also** section, this section reference itself.
 
  
#### Relevant Issues

 + Issue: [#982](https://github.com/arviz-devs/arviz/issues/2080): Unable to plot ppc.

### 3. Rank plots

#### Analysis of current defaults

 + The visualization is not clear when `kind = "vlines"`, an additional explanation of how to visualize convergence of chains when using this kind of plots will be needed.
 
 + This function has a significantly short functionality and diversity compared to the ppc methods provided by the bayesplot package.
 
#### Analysis of usage advise

  + The x-axis label confuses me. What does "ranks(all chains)" mean? Does the function mix all the simulated chains, calculate ranks, and then somehow separate them between chains? Or Are the ranks for the different chains? The label "ranks" in both cases is minimalistic and less confusing.

### 4. Separation plots

#### Analysis of current defaults

 + The plot's default height size is too wide.
 
#### Analysis of usage advise

  + The plot's documentation is not enough. 

### 5. Trace plots

#### Analysis of current defaults

 + The function does not provide a vanilla trace plot, but a combo of different plots.
 
 + This function has a significantly short functionality and diversity compared to the `mcmc_combo` function provided by the bayesplot package.
 
#### Analysis of usage advise

  + Set `kind="rank_bar"` options as default.
 
#### Relevant Issues

 + Issue: [#982](https://github.com/arviz-devs/arviz/issues/2080): Unable to plot ppc.

### 6. Time series plots

#### Analysis of current defaults

  + If an `x` value is not specified, add a default label to the `x.label`, like `"Data point (index)"` or "time".

### 8. Violin plots

#### Analysis of current defaults

  + The inner boxplot is too small, make it bigger for a clear visualization.
  
  + bayesplot provides an option to display a violin plot per chain, which is not available in ArviZ.
