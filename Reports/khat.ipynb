## Pareto $\hat k$ plots for LOO's diagnosing convergence.

The `plot_khat` function from `ArviZ` plots the estimated shape parameters $\hat k$ from the estimated elpd by a PSIS approximation.

This function is the ArviZ equivalent to the `plot()` methods from the [loo](http://mc-stan.org/loo/articles/loo2-example.html) package. The functionality of both functions are the same, but the `plot` method displays the threshold hlines by default, and `plot_khat` only displays them if its specified.

@fig-khat shows the calculated $\hat k$ of the LOO estimates for the model applied to the Radon data-set.As all values are lower than 0.7, we accept the PISIS approximation of the LOO-CV.

```python
#| label: fig-khat
#| fig-cap: "Kernel density estimated plots for visualizing the marginal posteriors of mu for the non-centered model, 8-schools dataset."

import numpy  as np
import arviz  as az
import matplotlib.pyplot as plt

az.style.use("arviz-darkgrid")

radon = az.load_arviz_data("radon")
loo_radon = az.loo(radon, pointwise=True)

az.plot_khat(loo_radon)
plt.show()
```

The pre-asymptotic convergence rate of importance sampling can be estimated based on the fractional number of finite moments of the importance ratio distribution. GPD is fitted to the largest importance ratios and interprets the estimated shape parameter $\hat{k}$ as a  diagnostic (most importantly, if $\hat{k} > 0.7$, then the convergence rate is impractically low).
