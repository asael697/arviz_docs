## Pair marginals plots for dispersion. 

The `plot_pair` function from `ArviZ` plots a grid of scatter, 2-dimensional densities, or hexbin plots, for dispersion evauation of multiple marginal posteriors.

This function is the ArviZ similar to the `mcmc_pairs()` function from the [bayesplot](http://mc-stan.org/bayesplot/reference/MCMC-scatterplots.html) package.

@fig-pair shows compares the model's posterior predictive samples with the original data.

```python
#| label: fig-pair
#| fig-cap: "Scatter plots for pairing the posteriors of mu and theta, 8-schools datasets; at Choate and Deerfield levels."

import arviz  as az
import matplotlib.pyplot as plt

az.style.use("arviz-darkgrid")
idata = az.load_arviz_data("centered_eight")

az.plot_pair(idata,var_names=['mu','theta'],coords = {'school': ['Choate', 'Deerfield']})
plt.show()
```
