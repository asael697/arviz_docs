## Posterior predictive check plots for time series data

The `plot_ts` function from `ArviZ` plots Posterior predictive checks and mean plots for time series like data.

This function is the ArviZ similar to the `ppc_ribbon()` and `ppc_intervals()` functions from the [bayesplot](http://mc-stan.org/bayesplot/reference/PPC-intervals.html) package.

@fig-ts shows compares the model's posterior predictive samples with the original data.

```python
#| label: fig-ts
#| fig-cap: "Ribbon ppc plot for evaluating a Gaussian white noise model."

import arviz as az
import numpy as np
import matplotlib.pyplot as plt

az.style.use("arviz-darkgrid")

nchains, ndraws = (4, 1000)

obs_data = {"y": -2 * np.random.normal(size = 100)}
posterior_predictive = {
    "y": np.random.normal(obs_data["y"]-np.random.normal(loc=0.5,size=1), 
                          size=(nchains, ndraws, len(obs_data["y"])))
 }

idata = az.from_dict(observed_data=obs_data,
    posterior_predictive=posterior_predictive)

ax = az.plot_ts(idata=idata, y="y")
```
