## High density interval plots for regression data

The `plot_hdi` function from `ArviZ` plots high predictive density intervals for regression data.

@fig-hdi shows the high density intervals, for a vanilla regression output, where `x1` simulates the independent variable, and `y` simulates the posterior_predictive samples of fitted regression model `y ~ b0 + b1*x1`, with two chains of 1,000 iterations, from a $n = 100$ dataset.

```python
#| warnings: false
#| label: fig-hid
#| fig-cap: " High predictive density intervals for the posterior predictive samples of `y ~ b0 + b1 x1` model"

import numpy  as np
import arviz  as az
import matplotlib.pyplot as plt

az.style.use("arviz-darkgrid")

# The unknown regression parameters
b = np.random.normal(0,1,2)
x1 = np.random.normal(0, 1, 100)

# Simulated pp samples form the regression model
y  = np.random.normal(b[0]+x1*b[1],1,size=(2, 1000, 100))

hdi_data = az.hdi(y)

az.plot_hdi(x1, hdi_data = hdi_data)
plt.show()
```
