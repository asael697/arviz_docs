## Bayesian p-values plot

The `plot_bpc()` function from [ArviZ](https://arviz-devs.github.io/arviz/index.html) displays the Bayesian p-values for observed data and posterior predictive checks. These plots' main functionality is for model diagnostics using predictive distribution. These methods are usually known as a posterior predictive check (PPC). The `ppc_stat()` function of [bayesplot](https://mc-stan.org/bayesplot/index.html) provides equivalent functionality to the `plot_bpc()` when setting the option `kind = "t_stat"`. Overall, there is no function [bayesplot](https://mc-stan.org/bayesplot/index.html) with the same functionality as [ArviZ's](https://arviz-devs.github.io/arviz/index.html) `plot_bpc()` function.

## Analysis of current defaults

@fig-uvalues shows the current defaults for the Bayesian p-values plots, where:

 + The left subplot performs a visual representation of a uniform test. If $y^*|y$ and $y$ have the same distribution, then the plot seems like a continuous uniform distribution ($U[0,1]$).
 
 + The centered subplot shows the posterior distribution of $\text{p-value} = P(y^* \leq y | y)$. If $y^*$ and $y$ have the same distribution, then the plot seems like a symmetric continuous distribution centered at 0.5.
 
 + Similarly, the right subplot shows the posterior distribution of $\text{p-value} = P(T(y^*) \leq T(y) | y)$, where $T()$ is a statistic of $y$. By default, the sample mean is the statistic shown.

```python
#| label: fig-uvalues
#| fig-cap: "Posterior predictive checks using p-values plot for the 'regression1d' dataset."

import matplotlib.pyplot as plt
import arviz as az

az.style.use("arviz-darkgrid")
data = az.load_arviz_data("regression1d")

_, ax = plt.subplots(1, 3, figsize=(10, 3))

az.plot_bpv(data,ax = ax[0])
az.plot_bpv(data,kind =  "p_value",ax = ax[1])
az.plot_bpv(data,kind = "t_stat",ax = ax[2])

ax[0].set_xlabel('u-value')
ax[1].set_xlabel('p-value')
ax[2].set_xlabel('t_stat')

plt.show()
```
