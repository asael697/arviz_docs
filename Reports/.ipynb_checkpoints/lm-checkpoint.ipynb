## Posterior predictive check plots for regression. data

The `plot_lm` function from `ArviZ` plots Posterior predictive checks and mean plots for regression-like data.

This function is the ArviZ similar to the `ppc_ribbon()` and `ppc_intervals()` functions from the [bayesplot](http://mc-stan.org/bayesplot/reference/PPC-intervals.html) package.

@fig-lm shows compares the model's posterior predictive samples with the original data.

```python
#| label: fig-lm
#| fig-cap: "Posterior Predictive checks for regression data."

import numpy  as np
import arviz  as az
import matplotlib.pyplot as plt

_, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

az.style.use("arviz-darkgrid")
idata = az.load_arviz_data('regression1d')

az.plot_lm(idata=idata,y="y",axes=ax1)
az.plot_lm(idata=idata,y="y",kind_pp="hdi",axes=ax2)

ax1.set_title("Posterior predicts: Samples")
ax2.set_title("Posterior predicts: Predictive intervals")
plt.show()
```
