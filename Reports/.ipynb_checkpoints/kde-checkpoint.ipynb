## kernel density estimated plots for visualizing posteriors

The `plot_kde` function from `ArviZ` plots densities using kernel estimations, for one and two dimensional random variables.

@fig-kde shows the estimated 2 dimensional densities for a multivariate normal distribution, where the left and right plots are the marginal densities, and the centered plot is the  bivariated  density.

```python
#| warnings: false
#| label: fig-kde
#| fig-cap: "Kernel density estimated plots for visualizing the marginal posteriors of mu for the non-centered model, 8-schools dataset."

import numpy  as np
import arviz  as az
import matplotlib.pyplot as plt

az.style.use("arviz-darkgrid")

non_centered = az.load_arviz_data('non_centered_eight')
mu_posterior = non_centered.posterior["mu"].values

_, ax = plt.subplots(1, 3, figsize=(10, 3.5))

az.plot_kde(mu_posterior[0,:],ax = ax[0])
az.plot_kde(mu_posterior[0,:],mu_posterior[1,:],ax = ax[1])
az.plot_kde(mu_posterior[1,:], rotated=True,ax = ax[2])

ax[0].set_xlabel('$\mu_0$')
ax[1].set_xlabel('$\mu_0$')
ax[1].set_ylabel('$\mu_1$')
ax[2].set_xlabel('$\mu_1$')

plt.show()
```
