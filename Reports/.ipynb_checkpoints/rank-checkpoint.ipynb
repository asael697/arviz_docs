## Rank plots

The `plot_rank` function from `ArviZ` plots histograms of the ranked posterior draws for each chain. This function is the ArviZ equivalent of `mcmc_rank_hist()` function from the [bayesplot](https://mc-stan.org/bayesplot/reference/MCMC-traces.html) package. The functionality and defaults between packages are similar.

@fig-rank shows the default rank plots for $\mu$ in the eight schools example.

```python
#| label: fig-rank
#| fig-cap: "Posterior Predictive checks, for the Radon dataset."

import arviz as az
import matplotlib.pyplot as plt

data = az.load_arviz_data('centered_eight')
az.style.use("arviz-darkgrid")

_, ax = plt.subplots(1, 2, figsize=(10, 3))

az.plot_rank(data,var_names="mu",ax = ax[0])
az.plot_rank(data,var_names="mu",kind='vlines',ax = ax[1])

plt.show()
```
