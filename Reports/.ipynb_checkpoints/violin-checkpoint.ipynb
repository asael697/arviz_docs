## Violin plots for posterior visualization

The `plot_violin` function from `ArviZ` helps visualize the posterior distribution using a kernel density estimation method (KDE). This function is the ArviZ equivalent of `mcmc_violin()` function from the [bayesplot](https://mc-stan.org/bayesplot/reference/MCMC-distributions.html) package. 

@fig-violin shows the default violin plots for $\mu$ and $\tau$ in the eight schools example.

```python
#| label: fig-violin
#| fig-cap: "Violin plots for mu and tau parameters in the eitght schools example
import arviz as az
import matplotlib.pyplot as plt

data = az.load_arviz_data('centered_eight')
az.style.use("arviz-darkgrid")

az.plot_violin(data,var_names=['mu','tau'])
plt.show()
```
