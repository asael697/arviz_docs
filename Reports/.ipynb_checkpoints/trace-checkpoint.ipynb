## Trace and Rank plots

The `plot_trace` function from `ArviZ` helps visualize the posterior distribution using a kernel density estimation method (KDE). This function is the ArviZ equivalent of `mcmc_combo()` function from the [bayesplot](https://mc-stan.org/bayesplot/reference/MCMC-combos.html) package. But ArviZ's function is limited to trace, ranks and density plots only, where `mcmc_combo` has more posible types of plots.

@fig-trace shows the default density plots for $\theta$ in the eight schools example.

```python
#| label: fig-trace
#| fig-cap: "Trace and Density plots of the Centered model for the Eight school dataset."

import arviz as az
import matplotlib.pyplot as plt

data = az.load_arviz_data('centered_eight')
az.style.use("arviz-darkgrid")

az.plot_trace(data, var_names=["mu", "tau"])
plt.show()
```
