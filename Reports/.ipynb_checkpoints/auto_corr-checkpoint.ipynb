## ACF plots

The `plot_autocorr()` function from [ArviZ](https://arviz-devs.github.io/arviz/index.html) displays the chain's auto correlation function (ACF). These plots main functionality are a diagnostic for chain convergence.  This function is the ArviZ equivalent of `mcmc_acf_bar()` function from the [bayesplot](https://mc-stan.org/bayesplot/index.html) `R` package. The functionality and defaults of the two functions are the same, providing a visualization of the chain's ACF.

```python
#| label: fig-ACF
#| fig-cap: "ACF plots for diagnostic convergence of mu and tau's chains."

import matplotlib.pyplot as plt
import arviz as az

az.style.use("arviz-darkgrid")
data = az.load_arviz_data('centered_eight')

az.plot_autocorr(data, var_names=['mu','tau'])
plt.show()
```

## Analysis of current defaults

The `plot_autocorr()` displays the acf using vertical bars for every parameter's posterior, and for every chain. The visualization is neat, and its major utility is for convergence diagnostics.

### Comments

 + Using vertical lines to display the ACF is the usual way to present these kinds of plots, contrary to the line plot used in bayesplot, which might be unclear.
 
 + The plot, by default, shows a dark confidence band representing the region where the ACF is zero. I consider this a good practice. 
 
 + The x-axis label is blank by default, could be changed to: "number of lags" or simply "lags."
 
 + The y-axis label is blank by default, could be changed to: "ACF."
