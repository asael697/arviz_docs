## Dot plots

The `plot_dot` function from `ArviZ` is an alternative method for visualizing distributions.

@fig-dot shows the default dot plots for different simulated samples using numpy

```python
#| label: fig-dist
#| fig-cap: "Density plot and histogram for simulated random samples"

import arviz as az
import matplotlib.pyplot as plt
import numpy as np

az.style.use("arviz-darkgrid")

_, ax = plt.subplots(1, 3, figsize=(10, 3))

a = np.random.normal(0, 1, 1000)
b = np.random.binomial(10,0.5,1000)
c = 2*a + a*b

az.plot_dot(a,ax=ax[0])
az.plot_dot(b,ax=ax[1])
az.plot_dot(c,ax=ax[2])

ax[0].set_xlabel('Dots N(0,1)')
ax[1].set_xlabel('Dots Binom(10;0.5)')
ax[2].set_xlabel('Dots unknown sample')

plt.show()
```

## Analysis of current defaults

The dot plots display a natural a neat visualization for a simulated random sample. Avoiding the natural problems of selecting the number of bins on histrograms.
