## PPC plots

The `plot_ppc` function from `ArviZ` helps check model using the posterior predictive density with tree different methods:

  + Overlayed densities  (`kind = "kde"`).
  
  + Overlayed ecdfs (`kind = "cumulative"`)
  
  + Simultaneous overlayed densities and scatter plots(`kind = "scatter"`)
  
This function is the ArviZ equivalent of `ppc_dens_overlay()`, `ppc_dens_overlay_grouped()`, `ppc_ecdf_overlay()`, `ppc_ecdf_overlay_grouped()`, `ppc_dens()`,  and`ppc_hist` functions from the [bayesplot](http://mc-stan.org/bayesplot/reference/PPC-distributions.html) package. The functionality and defaults between packages is similar.

@fig-dens shows the default density plots for $\theta$ in the eight schools example.

```python
#| label: fig-ppc
#| fig-cap: "Posterior Predictive checks, for the Radon dataset."

import arviz as az
import matplotlib.pyplot as plt

data = az.load_arviz_data('radon')
az.style.use("arviz-darkgrid")

_, ax = plt.subplots(1, 3, figsize=(10, 3))

az.plot_ppc(data,kind="kde",ax = ax[0])
az.plot_ppc(data,kind="cumulative",ax = ax[1])
az.plot_ppc(data,kind="scatter",ax = ax[2])

plt.show()
```

     - Plot for different posterior/prior predictive checks:
     
         * The `kind="kde"` argument generates a overlay density plot, each simulated predictive distribution is overlaid with the density of data itself. The densities are estimated using a kernel density method.
         
         * The `kind="ecdf"` argument generates a overlay ecdf plot, the ecdf of each simulated predictive distribution is overlaid with the ecdf of data itself. 
     
         * The `kind="scatter"` argument generates a plot that overlays using a stacked scatter plot, this is particularly useful when the sample sizes are small.
