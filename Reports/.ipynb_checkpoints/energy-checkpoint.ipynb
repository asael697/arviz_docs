## Energy plots

The `plot_energy` function from `ArviZ` displays gradient-based MCMC energy transition distribution and marginal energy distribution plots.

@fig-nuts shows the energy transition and marginal energy distributions for the centered model, 8-schools data set.

```python
#| label: fig-nuts
#| fig-cap: "Density plots for the energy transition and marginal energy distributions, 8-schools centered model"

import arviz as az
import matplotlib.pyplot as plt

data = az.load_arviz_data('centered_eight')
az.plot_energy(data)
plt.show()
```
