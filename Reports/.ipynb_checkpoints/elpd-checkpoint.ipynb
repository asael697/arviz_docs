## elpd plots

The `plot_ecdf` function from `ArviZ` is an alternative method for visualizing distributions.

@fig-elpd shows a elpd comparison of two hierachical models for the 8-schools dataset. 

```python
#| warnings: false
#| label: fig-polar
#| fig-cap: "Pointwise elpd comparison of two models (centered and non-centered) for the 8-schools example"

import arviz as az
import matplotlib.pyplot as plt

idata1 = az.load_arviz_data("centered_eight")
idata2 = az.load_arviz_data("non_centered_eight")

dict1 = {"centered model": idata1, 
         "non centered model": idata2}

az.plot_elpd(dict1)
plt.show()
```
