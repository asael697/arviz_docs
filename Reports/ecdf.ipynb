## ecdf plots

The `plot_ecdf` function from `ArviZ` is an alternative method for visualizing distributions.

@fig-ecdf shows the cummulative function of a simulated sample from a standard normal distribution. 

```python
import arviz as az
import matplotlib.pyplot as plt

from scipy.stats import  norm, binom

az.style.use("arviz-darkgrid")
_, ax = plt.subplots(2, 2, figsize=(10, 3))

## Sample and distribution 
sample = norm(0,1).rvs(1000)
dist = norm(0,1)


az.plot_ecdf(sample,ax = ax[0,0])
az.plot_ecdf(sample,cdf=dist.cdf,ax = ax[0,1])

az.plot_ecdf(sample, cdf = dist.cdf,
            confidence_bands = True, 
            difference = True,ax=ax[1,0])

az.plot_ecdf(sample, cdf = dist.cdf,
             confidence_bands = True, 
             pit = True,ax = ax[1,1])

ax[0,0].set_xlabel('Default ecdf')
ax[0,1].set_xlabel('ecdf compare: sample vs N(0,1)')
ax[1,0].set_xlabel('Differenced ecdf compare')
ax[1,1].set_xlabel('PIT: sample vs N(0,1)')

plt.show()
```
