## Monte Carlo Standard Error plots 

The `plot_mcse` function from `ArviZ` plots the local Monte Carlo Standard Error for the quantile estimates of an specified posterior.

@fig-mcse shows compares the model's posterior predictive samples with the original data.

```python
#| label: fig-mcse
#| fig-cap: "Monte Carlo Standard Errors plot for the quantile estimates of mu, 8-schools data sets."

import arviz  as az
import matplotlib.pyplot as plt

az.style.use("arviz-darkgrid")
idata = az.load_arviz_data("centered_eight")

az.plot_mcse(idata, var_names=["mu"])
plt.show()
```
