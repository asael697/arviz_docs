## Density plots

The `plot_density` function from `ArviZ` helps visualize the posterior distribution using a kernel density estimation method (KDE). This function is the ArviZ equivalent of `mcmc_hist()` and `mcmc_dens()` functions from the [bayesplot](https://mc-stan.org/bayesplot/reference/MCMC-intervals.html) package. The functionality and defaults are similar for the three functions, but `ArviZ` plot,s by default. present the distribution truncated at their $100\times(1-\alpha)\%$ highest density intervals.

@fig-dens shows the default density plots for $\theta$ in the eight schools example.

```python
#| label: fig-dens
#| fig-cap: "Posterior densities for theta, 8 schools example"

import arviz as az
import matplotlib.pyplot as plt

centered = az.load_arviz_data('centered_eight')
az.style.use("arviz-darkgrid")

az.plot_density(centered)

plt.show()
```

## Analysis of current defaults

The density plots display credible intervals and densities for every parameter's posterior. These plots are commonly used for visualizing posteriors. ArviZ plots improve bayesplot by offering posterior visualization for diagnostic and a naive visualization of the parameter's credible intervals.
