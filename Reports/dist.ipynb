## Distribution plots

The `plot_dist` function from `ArviZ` helps visualize distributions using histograms and kernel density estimation method (KDE). 

@fig-dist shows the default density plots for different simulated samples using numpy.

```python
#| label: fig-dist
#| fig-cap: "Density plot and histogram for simulated random samples"

import arviz as az
import matplotlib.pyplot as plt
import numpy as np

az.style.use("arviz-darkgrid")

_, ax = plt.subplots(1, 3, figsize=(10, 3))

a = np.random.normal(0, 1, 1000)
b = np.random.binomial(10,0.5,1000)
c = 2*a + a*b

az.plot_dist(a,ax=ax[0])
az.plot_dist(b,ax=ax[1])
az.plot_dist(c,ax=ax[2])

ax[0].set_xlabel('Density N(0,1)')
ax[1].set_xlabel('Histogram Binom(10;0.5)')
ax[2].set_xlabel('Density unknown sample')

plt.show()
```
