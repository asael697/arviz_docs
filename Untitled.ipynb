## Introduction

The `plot_compare()` function from `ArviZ` helps visualize multiple posteriors. Currently, there is no equivalent function in the [bayesplot](https://mc-stan.org/bayesplot/reference/MCMC-intervals.html) package. 

@fig-polar shows the current default forest and ridge plots, for $\theta$ in the eight schools example.

```python
#| label: fig-ex-comb
#| fig-cap: "Posterior estimates for theta, 8 schools example, chains combined"

import arviz as az

az.style.use("arviz-darkgrid")

model_compare = az.compare({'Centered 8 schools': az.load_arviz_data('centered_eight'),
                            'Non-centered 8 schools': az.load_arviz_data('non_centered_eight')})

az.plot_compare(model_compare)
```
